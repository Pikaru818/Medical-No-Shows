Step1/2:

Adds the dataset and removes NaN values and copies the dataframe for datachecking and confirmation

```python
import pandas as pd
import numpy as np
import sklearn.datasets 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder,StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
sca = StandardScaler()
enc = LabelEncoder()
oenc = OneHotEncoder()
```

```python
df = pd.read_csv("Medical.csv")
```

```python
df=df.replace(r'^\s*$', np.nan, regex=True)
df.dropna(inplace=True,axis=0)
df
```

```python
mdf= df.copy()
```

Step 3:

Extracts the features

```python
mdf=mdf.drop(columns=['PatientId',"AppointmentID","ScheduledDay","AppointmentDay","Neighbourhood"])
```

```python
mdf
```

Step 4:

Encodes the features that are not integers

```python
No_show= mdf.iloc[:, 8]

mdf['Gender']= enc.fit_transform(mdf.Gender.values)
mdf['No-show']= enc.fit_transform(No_show.values)
```

```python
mdf
```

Step 5:

Splits the data

```python
X= mdf.iloc[:, 0:8].values
y=mdf.iloc[:, 8].values
```

```python
from sklearn.model_selection import train_test_split
train_ratio = 0.80
validation_ratio = 0.10
test_ratio = 0.10

# train is now 80% of the entire data set
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=1 - train_ratio)

# test is now 10% of the initial data set
# validation is now 10% of the initial data set
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=test_ratio/(test_ratio + validation_ratio)) 

print(x_train, x_val, x_test)
```

Step 6.1:

Performs DecisionTreeClassifier and ends up with a decent result of 79$ on both the validation and test scores and prints them off

```python
from sklearn.tree import DecisionTreeClassifier
Dmodel = DecisionTreeClassifier(criterion="entropy" , splitter='best')
Dmodel.fit(x_train, y_train)
D_pred = Dmodel.predict(x_test)
Dscore = Dmodel.score(x_val,y_val)
from sklearn.metrics import recall_score
print('Score using the test data: '+ str(recall_score(y_test, D_pred,  average='weighted')))
print('Score using the validation data: ',Dscore)
```

Step 6.2:

Prints the confusion matrix and classification report

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test, D_pred))
print(classification_report(y_test, D_pred))
```

Step 7:

Performs a RandomForestClassifier and ends up with a very jank variant where the test data never breaches 2% but the validation is always 79% so honestly I would prefer to use a Decision Tree Classifier in this case

```python
from sklearn.ensemble import RandomForestClassifier 
Rmodel = RandomForestClassifier(n_estimators=10, criterion='entropy')
Rmodel.fit(x_train, y_train)
    
R_pred = Rmodel.predict(x_test)
    
Rscore = Rmodel.score(x_val,y_val)
    
from sklearn.metrics import recall_score
print('Score using the test data: ' + str(recall_score(y_test, R_pred)))
print('Score using the validation data: ',Rscore)
```
